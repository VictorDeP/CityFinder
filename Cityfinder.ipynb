```python
import re
import csv
```

```python
# Get list of lists from csv of all world cities with more than 15000 population

with open('world-cities.csv', newline='',encoding="utf8") as f:
    reader = reader = csv.reader(f)
    list_reader = list(reader)
```

```python
#Country limitation

selected_countries = ['Belgium', 'Luxembourg', 'Germany', 'France', 'The Netherlands', 'United Kingdom', 'Monaco', 'Switzerland', 'Spain', 'Italy', 'Denmark']

country_limited_list_reader = [x for x in list_reader if x[1] in selected_countries]

print(len(country_limited_list_reader))
print (country_limited_list_reader)
```

```python
#Limit cities containing all of selected letters
selected_letters = ['w', 'm', 'b', 's', 'l']

letters_limited_list_reader = [x for x in country_limited_list_reader if all(l.lower() in x[0].lower() for l in selected_letters)]

print(len(letters_limited_list_reader))
print(letters_limited_list_reader)
```
